Q1:What is the relationship between a player's first serve percentage and the number of aces they hit in the Australian Open Men's 2013 tournament?


```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# load the data
df = pd.read_csv('/content/AusOpen-men-2013.csv')

# calculate the correlation between FSP and ACE
corr = df['FSP.1'].corr(df['ACE.1'])

# plot the scatter plot
plt.scatter(df['FSP.1'], df['ACE.1'])
plt.xlabel('First Serve Percentage')
plt.ylabel('Number of Aces')
plt.title('Relationship between First Serve Percentage and Number of Aces')

# fit the linear regression model
X = df[['FSP.1']]
y = df['ACE.1']
reg = LinearRegression().fit(X, y)

# plot the regression line
plt.plot(X, reg.predict(X), color='red', linewidth=2)

plt.show()

```

Q2:What is the impact of the first serve percentage (FSP) on the result of a tennis match?


```python
def get_win_count(df1):
  df1['Favourite'] = df1.apply(lambda x:1 if x['FSP.1']>x['FSP.2'] else 0, axis=1)
  favourable_win_1 = df1.loc[df1['Result']==df1['Favourite']].shape[0]
  unfavourable_win_1 = df1.loc[df1['Result']!=df1['Favourite']].shape[0]
  return favourable_win_1,unfavourable_win_1

win_count_list=[]
for data in all_dataframe:
  a,b = get_win_count(data)
  win_count_list.append([a,b])

total_favourable_win = 0
total_unfavourable_win = 0
for i in range(8):
  total_favourable_win += win_count_list[i][0]
for i in range(8):

  total_unfavourable_win += win_count_list[i][1]


plt.figure(figsize=(6,6))
labels = ['Player with higher FSP wins', 'Player with lesser FSP wins']
sizes = [total_favourable_win, total_unfavourable_win]
colors = ['#66b3ff','#ff6666']
sns.set_style("whitegrid")
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90)
plt.title('Impact of the FSP on the result of a match')
plt.show()
```

Q3:How does the ratio of winners to unforced errors of the Wimbledon 2013 finalists  (men and women) compare, and how does it impact the result?


```python
import pandas as pd
import matplotlib.pyplot as plt

# load the data
df = pd.read_csv('/content/Wimbledon-men-2013.csv')

# filter for round 7 matches
df_round7 = df[df['Round'] == 7]

# calculate the golden ratio for player 1 and player 2
df_round7['GoldenRatio1'] = df_round7['WNR.1'] / df_round7['UFE.1']
df_round7['GoldenRatio2'] = df_round7['WNR.2'] / df_round7['UFE.2']

# plot the golden ratio for player 1 and player 2
plt.plot(df_round7['Player1'], df_round7['GoldenRatio1'], 'bo-', label='Player 1')
plt.plot(df_round7['Player2'], df_round7['GoldenRatio2'], 'go-', label='Player 2')
plt.xticks(rotation=90)
plt.xlabel('Player')
plt.ylabel('Golden Ratio')
plt.legend()
plt.grid(True)
plt.show()

# load the data
df = pd.read_csv('/content/Wimbledon-women-2013.csv')

# filter for round 7 matches
df_round7 = df[df['Round'] == 7]

# calculate the golden ratio for player 1 and player 2
df_round7['GoldenRatio1'] = df_round7['WNR.1'] / df_round7['UFE.1']
df_round7['GoldenRatio2'] = df_round7['WNR.2'] / df_round7['UFE.2']

# plot the golden ratio for player 1 and player 2
plt.plot(df_round7['Player1'], df_round7['GoldenRatio1'], 'bo-', label='Player 1')
plt.plot(df_round7['Player2'], df_round7['GoldenRatio2'], 'go-', label='Player 2')
plt.xticks(rotation=90)
plt.xlabel('Player')
plt.ylabel('Golden Ratio')
plt.legend()
plt.grid(True)
plt.show()
```

Q8:How do the different parameters compare for the French Open 2013 semi-finalists?

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
df3 = pd.read_csv("/content/FrenchOpen-men-2013.csv")

# Select the relevant columns for the given players and their first serve percentages
df_player1 = df3[df3['Player1'].isin(['David Ferrer', 'Jo-Wilfried Tsonga', 'Novak Djokovic', 'Rafael Nadal'])][['Player1', 'Round', 'FSW.1']]
df_player2 = df3[df3['Player2'].isin(['David Ferrer', 'Jo-Wilfried Tsonga', 'Novak Djokovic', 'Rafael Nadal'])][['Player2', 'Round', 'FSW.2']]
df_player1.columns = ['Player', 'Round', 'FSW']
df_player2.columns = ['Player', 'Round', 'FSW']
df_first_serve_pct = pd.concat([df_player1, df_player2], ignore_index=True)

# Group the data by player and compute the mean of first serve percentage
grouped = df_first_serve_pct.groupby(['Player'])['FSW'].mean().reset_index()

# Sort the data by first serve percentage and plot the mean first serve percentage for each player
fig, ax = plt.subplots(figsize=(10, 6))
grouped = grouped.sort_values(by='FSW')
ax.barh(grouped['Player'], grouped['FSW'], color=['red', 'blue', 'green', 'purple'])
ax.axvline(grouped['FSW'].mean(), color='black', linestyle='--', label='Total Mean')
ax.set_xlabel('First Serve Percentage')
ax.set_ylabel('Player')
ax.set_title("Mean First Serve Percentage of Players in French Open Men's 2013 Tournament")
ax.legend()

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
df3 = pd.read_csv("/content/FrenchOpen-men-2013.csv")

# Select the relevant columns for the given players and their break points won
df_player1 = df3[df3['Player1'].isin(['David Ferrer', 'Jo-Wilfried Tsonga', 'Novak Djokovic', 'Rafael Nadal'])][['Player1', 'BPW.1']]
df_player2 = df3[df3['Player2'].isin(['David Ferrer', 'Jo-Wilfried Tsonga', 'Novak Djokovic', 'Rafael Nadal'])][['Player2', 'BPW.2']]
df_player1.columns = ['Player', 'BPW']
df_player2.columns = ['Player', 'BPW']
df_break_points_won = pd.concat([df_player1, df_player2], ignore_index=True)

# Compute the mean break points won for each player
mean_BPW = df_break_points_won.groupby(['Player'])['BPW'].mean().reset_index()
mean_BPW_sorted = mean_BPW.sort_values(by='BPW', ascending=True)

# Compute the overall mean of the "BPW" column
overall_mean_BPW = df_break_points_won['BPW'].mean()

# Create a horizontal bar graph of the mean break points won for each player and add a vertical line for the overall mean
fig, ax = plt.subplots(figsize=(10, 6))
ax.barh(mean_BPW_sorted['Player'], mean_BPW_sorted['BPW'], color=['red', 'blue', 'green', 'purple'])
ax.axvline(x=overall_mean_BPW, color='black', linestyle='--', label='Overall Mean')

ax.set_xlabel('Mean Number of Break Points Won')
ax.set_ylabel('Player')
ax.set_title("Mean Number of Break Points Won by Players in French Open Men's 2013 Tournament")
ax.legend()

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
df3 = pd.read_csv("/content/FrenchOpen-men-2013.csv")

# Select the relevant columns for the given players and their winners hit
df_player1 = df3[df3['Player1'].isin(['David Ferrer', 'Jo-Wilfried Tsonga', 'Novak Djokovic', 'Rafael Nadal'])][['Player1', 'WNR.1']]
df_player2 = df3[df3['Player2'].isin(['David Ferrer', 'Jo-Wilfried Tsonga', 'Novak Djokovic', 'Rafael Nadal'])][['Player2', 'WNR.2']]
df_player1.columns = ['Player', 'WNR']
df_player2.columns = ['Player', 'WNR']
df_winners = pd.concat([df_player1, df_player2], ignore_index=True)

# Compute the mean of winners hit for each player
grouped = df_winners.groupby(['Player'])['WNR'].mean().reset_index()

# Sort the data in increasing order of mean winners hit
grouped = grouped.sort_values(by=['WNR'])
mean_WNR = df_winners['WNR'].mean()

# Create a horizontal bar graph for each player
fig, ax = plt.subplots(figsize=(10, 6))
players = list(grouped['Player'])
colors = ['red', 'blue', 'green', 'purple']
ax.barh(players, grouped['WNR'], color=colors)

ax.set_xlabel('Mean Number of Winners Hit')
ax.set_ylabel('Player')
ax.set_title("Mean Number of Winners Hit by Players in French Open Men's 2013 Tournament")
plt.axvline(x=mean_WNR, color='black', linestyle='--',label="Mean Winner's hit")
plt.legend()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
df3 = pd.read_csv("/content/FrenchOpen-men-2013.csv")

# Select the relevant columns for the given players and their aces
df_player1 = df3[df3['Player1'].isin(['David Ferrer', 'Jo-Wilfried Tsonga', 'Novak Djokovic', 'Rafael Nadal'])][['Player1', 'ACE.1']]
df_player2 = df3[df3['Player2'].isin(['David Ferrer', 'Jo-Wilfried Tsonga', 'Novak Djokovic', 'Rafael Nadal'])][['Player2', 'ACE.2']]
df_player1.columns = ['Player', 'ACE']
df_player2.columns = ['Player', 'ACE']
df_aces = pd.concat([df_player1, df_player2], ignore_index=True)

# Group the data by player and compute the mean of aces hit
grouped = df_aces.groupby('Player')['ACE'].mean().reset_index()

# Sort the data by mean aces hit in ascending order
grouped = grouped.sort_values(by='ACE')

# Compute the total mean of the mean aces hit values
total_mean = grouped['ACE'].mean()

# Create a horizontal bar chart for each player
fig, ax = plt.subplots(figsize=(10, 6))
players = grouped['Player']
aces = grouped['ACE']
colors = ['red', 'blue', 'green', 'purple']
ax.barh(players, aces, color=colors)

ax.set_xlabel('Mean Number of Aces')
ax.set_title("Mean Number of Aces Made by Players in French Open Men's 2013 Tournament")

# Add a vertical line for the total mean
ax.axvline(x=total_mean, color='black', linestyle='--')

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
df3 = pd.read_csv("/content/FrenchOpen-men-2013.csv")

# Select the relevant columns for the given players and their unforced errors
df_player1 = df3[df3['Player1'].isin(['David Ferrer', 'Jo-Wilfried Tsonga', 'Novak Djokovic', 'Rafael Nadal'])][['Player1', 'Round', 'UFE.1']]
df_player2 = df3[df3['Player2'].isin(['David Ferrer', 'Jo-Wilfried Tsonga', 'Novak Djokovic', 'Rafael Nadal'])][['Player2', 'Round', 'UFE.2']]
df_player1.columns = ['Player', 'Round', 'UFE']
df_player2.columns = ['Player', 'Round', 'UFE']
df_unforced_errors = pd.concat([df_player1, df_player2], ignore_index=True)

# Group the data by player and round number and compute the mean of unforced errors
grouped = df_unforced_errors.groupby(['Player'])['UFE'].mean().reset_index()

# Sort the data by mean unforced errors in ascending order
grouped = grouped.sort_values(by='UFE')

# Create a bar graph for each player
fig, ax = plt.subplots(figsize=(10, 6))
players = ['David Ferrer', 'Jo-Wilfried Tsonga', 'Novak Djokovic', 'Rafael Nadal']
colors = ['red', 'blue', 'green', 'purple']
y_pos = range(len(players))
ax.barh(y_pos, grouped['UFE'], color=colors)
ax.set_yticks(y_pos)
ax.set_yticklabels(players)
ax.invert_yaxis()  # labels read top-to-bottom
ax.set_xlabel('Mean Number of Unforced Errors')
ax.set_title("Mean Number of Unforced Errors Made by Players in French Open Men's 2013 Tournament")

# Add a vertical line for the total mean
total_mean = grouped['UFE'].mean()
ax.axvline(x=total_mean, color='black', linestyle='--')
ax.text(total_mean+0.1, 3.5, f'Total Mean: {total_mean:.2f}', va='center')
plt.show()
```
